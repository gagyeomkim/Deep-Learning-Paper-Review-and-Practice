## pre-trained 영어 BERT

```python
!pip install transformers
```

```python
from transformers import BertForMaskedLM
from transformers import AutoTokenizer
```

```python
# model과 tokenizer는 항상 사용하는 모델이 같아야함(맵핑관계)
model = BertForMaskedLM.from_pretrained('bert-large-uncased')
tokenizer = AutoTokenizer.from_pretrained("bert-large-uncased")
```

- `input_ids`: 정수 인코딩 결과 확인 가능
- `token_type_ids`: Segment 인코딩 결과 확인 가능
- `attention_mask`: 실제 단어와 패딩 토큰 구분

```python
inputs = tokenizer('Soccer is a really fun [MASK].')
print(f"정수 인코딩 결과 : {inputs['input_ids']}")
print(f"Segment 인코딩 결과 : {inputs['token_type_ids']}")
print(f"어텐션 마스크: {inputs['attention_mask']}")
```

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('Soccer is a really fun [MASK].')
```

```python
pip('The Avengers is a really fun [MASK].')
```

```python
pip('I went to [MASK] this morning.')
```

## pre-trained 한국어 BERT

```python
from transformers import BertForMaskedLM
from transformers import AutoTokenizer
```

```python
model = BertForMaskedLM.from_pretrained('klue/bert-base')
tokenizer = AutoTokenizer.from_pretrained('klue/bert-base')
```

```python
inputs = tokenizer('축구는 정말 재미있는 [MASK]다.')
print(f"정수 인코딩 결과 : {inputs['input_ids']}")
print(f"Segment 인코딩 결과 : {inputs['token_type_ids']}")
print(f"어텐션 마스크: {inputs['attention_mask']}")
```

```python
from transformers import FillMaskPipeline
pip = FillMaskPipeline(model=model, tokenizer=tokenizer)
```

```python
pip('축구는 정말 재미있는 [MASK]다.')
```

```python
pip('어벤져스는 정말 재미있는 [MASK]다.')
```

```python
pip('나는 오늘 아침에 [MASK]에 출근을 했다.')
```
